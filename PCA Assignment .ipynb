```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
import warnings
warnings.filterwarnings('ignore')
```

```python
wine=pd.read_csv('C:\\Users\\91741\\Downloads\\wine.csv')
```

```python
wine.head()
```

```python
wine.describe()
```

```python
wine.data=wine.iloc[:,1:]
wine.data.head()
```

```python
# Normalizing the numerical data 
wine_normal=scale(wine.data)
wine_normal
```

```python
#Using the PCA function of sklearn library and save in pca var upto PC1-PC3 
pca=PCA(n_components=3)
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
# The amount of variance that each PCA explains is 
var=pca.explained_variance_ratio_
var
```

```python
#Showing result of PC1
pca.components_[0]
```

```python
# Cumulative variance
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
#Variance plot for PCA components obtained 
plt.figure(figsize=(10,4))
plt.plot(var1,color='blue')
```

```python
# plot between PCA1 and PCA2
x=pca_values[:,0]
y=pca_values[:,1]
plt.scatter(x,y,color=['red'])
```

```python
#Clustering  

new_df=pd.DataFrame(pca_values[:,0:7])

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(new_df)
kmeans.labels_
```
